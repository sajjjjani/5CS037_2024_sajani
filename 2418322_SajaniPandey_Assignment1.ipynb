Task - 3 - Outlier Detection:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the South Asian dataset
dataset = pd.read_csv('/content/south_asian_dataset.csv')

# 1. Identify outlier countries in South Asia based on their Score and GDP per Capita
def detect_outliers(data, column):
    Q1 = data[column].quantile(0.25)  # First quartile (25%)
    Q3 = data[column].quantile(0.75)  # Third quartile (75%)
    IQR = Q3 - Q1                     # Interquartile range
    lower_bound = Q1 - 1.5 * IQR      # Lower threshold
    upper_bound = Q3 + 1.5 * IQR      # Upper threshold
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Detect outliers in 'Score' and 'GDP per Capita'
score_outliers, score_lower, score_upper = detect_outliers(dataset, 'score')
gdp_outliers, gdp_lower, gdp_upper = detect_outliers(dataset, 'Log GDP per capita')

# Combine outlier countries from both criteria
combined_outliers = pd.concat([score_outliers, gdp_outliers]).drop_duplicates()

print("Outlier countries based on Score:")
print(score_outliers[['Country name', 'score']])
print("\nOutlier countries based on GDP per Capita:")
print(gdp_outliers[['Country name', 'Log GDP per capita']])
print("\nCombined outliers:")
print(combined_outliers[['Country name', 'score', 'Log GDP per capita']])

# 2. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis
plt.figure(figsize=(8, 5))

# Scatter plot for all countries
plt.scatter(dataset['Log GDP per capita'], dataset['score'], label='Normal Countries', color='blue', alpha=0.7)

# Highlight outliers
plt.scatter(combined_outliers['Log GDP per capita'], combined_outliers['score'], label='Outliers', color='red', edgecolors='black', s=100)

# Annotate outliers
for _, row in combined_outliers.iterrows():
    plt.annotate(row['Country name'], (row['Log GDP per capita'], row['score']),
                 textcoords="offset points", xytext=(5, 5), ha='center', fontsize=9, color='darkred')

plt.xlabel('Log GDP per Capita')
plt.ylabel('score')
plt.title('Outlier Detection: GDP per Capita vs Score')
plt.legend()
plt.grid()
plt.show()

# 3. Discuss characteristics of outliers
# Calculate regional averages excluding outliers
regional_avg_score = dataset[~dataset['Country name'].isin(combined_outliers['Country name'])]['score'].mean()
regional_avg_gdp = dataset[~dataset['Country name'].isin(combined_outliers['Country name'])]['Log GDP per capita'].mean()

print(f"Regional Average Score (Excluding Outliers): {regional_avg_score:.2f}")
print(f"Regional Average Log GDP per Capita (Excluding Outliers): {regional_avg_gdp:.2f}")

```

Task - 4 - Exploring Trends Across Metrics:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the South Asian dataset
dataset = pd.read_csv('/content/south_asian_dataset.csv')

# 1. Choose two metrics and calculate their correlation with the Score
# Metrics: Freedom to Make Life Choices and Generosity
metrics = ['Freedom to make life choices', 'Generosity']

# Calculate Pearson correlation
correlations = {}
for metric in metrics:
    correlations[metric] = dataset['score'].corr(dataset[metric], method='pearson')

# Display correlations
print("Pearson Correlation with Score:")
for metric, corr_value in correlations.items():
    print(f"{metric}: {corr_value:.2f}")

# 2. Create scatter plots with trendlines and adjust for better clarity
plt.figure(figsize=(10, 5))

for i, metric in enumerate(metrics):
    plt.subplot(1, 2, i + 1)
    sns.regplot(x=dataset[metric], y=dataset['score'], scatter_kws={'color': 'blue', 's': 50},
                line_kws={'color': 'red', 'lw': 2})
    plt.xlabel(metric, fontsize=12)
    plt.ylabel('Score', fontsize=12)
    plt.title(f'Score vs {metric} (Correlation: {correlations[metric]:.2f})', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# 3. Identify and discuss strongest and weakest relationships
strongest_metric = max(correlations, key=correlations.get)
weakest_metric = min(correlations, key=correlations.get)

print(f"\nStrongest Relationship: {strongest_metric} with a correlation of {correlations[strongest_metric]:.2f}")
print(f"Weakest Relationship: {weakest_metric} with a correlation of {correlations[weakest_metric]:.2f}")

```

Task - 5 - Gap Analysis:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the South Asian dataset
dataset = pd.read_csv('/content/south_asian_dataset.csv')

# 1. Add a new column: GDP-Score Gap
dataset['GDP_Score_Gap'] = dataset['Log GDP per capita'] - dataset['score']

# 2. Rank countries by GDP-Score Gap in ascending and descending order
ascending_gap = dataset.sort_values(by='GDP_Score_Gap', ascending=True)
descending_gap = dataset.sort_values(by='GDP_Score_Gap', ascending=False)

# Display rankings
print("Countries Ranked by GDP-Score Gap (Ascending):")
print(ascending_gap[['Country name', 'GDP_Score_Gap']].head(3))

print("\nCountries Ranked by GDP-Score Gap (Descending):")
print(descending_gap[['Country name', 'GDP_Score_Gap']].head(3))

# 3. Highlight the top 3 countries with largest positive and negative gaps using a bar chart
top_positive_gaps = descending_gap.head(3)
top_negative_gaps = ascending_gap.head(3)

# Remove overlap to ensure a country does not appear in both lists
top_negative_gaps = top_negative_gaps[~top_negative_gaps['Country name'].isin(top_positive_gaps['Country name'])]

# Combine for visualization
gap_data = pd.concat([top_positive_gaps, top_negative_gaps])

# Bar chart with distinct colors for positive and negative gaps
plt.figure(figsize=(8, 5))
colors = ['blue' if gap > 0 else 'red' for gap in gap_data['GDP_Score_Gap']]
bars = plt.bar(gap_data['Country name'], gap_data['GDP_Score_Gap'], color=colors)

# Adding labels and gridlines
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.xlabel('Country Name')
plt.ylabel('GDP-Score Gap')
plt.title('Top 3 Positive and Negative GDP-Score Gaps')
plt.xticks(rotation=45)

# Adding bar annotations
for bar in bars:
    height = bar.get_height()
    label_position = height + 0.1 if height > 0 else height - 0.2
    plt.text(bar.get_x() + bar.get_width() / 2, label_position, f"{height:.2f}",
             ha='center', va='center', fontsize=9, color='black')

# Display gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# 4. Analysis of reasons behind these gaps
print("\nAnalysis of GDP-Score Gaps:")
for _, row in gap_data.iterrows():
    if row['GDP_Score_Gap'] > 0:
        print(f"{row['Country name']}: Positive gap suggests high GDP per capita relative to perceived well-being (Score).")
    else:
        print(f"{row['Country name']}: Negative gap suggests perceived well-being (Score) exceeds expectations based on GDP per capita.")
```

4. Happiness Disparity:

• Compute the range (max - min) and coefficient of variation (CV) for Score in both regions.

• Which region has greater variability in happiness?

```python
import pandas as pd

# Load the dataset
middle_east_data = pd.read_csv('/content/middle_eastern_dataset.csv')
south_asia_data = pd.read_csv('/content/south_asian_dataset.csv')

# 1. Calculate Range (Max - Min) for Score in both regions
south_asia_range = south_asia_data['score'].max() - south_asia_data['score'].min()
middle_east_range = middle_east_data['score'].max() - middle_east_data['score'].min()

# 2. Calculate Coefficient of Variation (CV) for Score in both regions
south_asia_mean = south_asia_data['score'].mean()
south_asia_std = south_asia_data['score'].std()
south_asia_cv = south_asia_std / south_asia_mean

middle_east_mean = middle_east_data['score'].mean()
middle_east_std = middle_east_data['score'].std()
middle_east_cv = middle_east_std / middle_east_mean

# Display the results
print("Happiness Disparity Analysis:")
print(f"South Asia Range: {south_asia_range}")
print(f"Middle East Range: {middle_east_range}")
print(f"South Asia Coefficient of Variation (CV): {south_asia_cv:.2f}")
print(f"Middle East Coefficient of Variation (CV): {middle_east_cv:.2f}")

# Conclusion on which region has greater variability in happiness
if south_asia_cv > middle_east_cv:
    print("\nSouth Asia has greater variability in happiness.")
else:
    print("\nMiddle East has greater variability in happiness.")
```

6. Outlier Detection:

• Identify outlier countries in both regions based on Score and GDP per Capita.

• Plot these outliers and discuss their implications.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
south_asia_data = pd.read_csv('/content/south_asian_dataset.csv')
middle_east_data = pd.read_csv('/content/middle_eastern_dataset.csv')

# Function to detect outliers using 1.5 × IQR rule
def detect_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Detect outliers for Score and GDP per Capita in both regions
south_asia_outliers_score = detect_outliers(south_asia_data, 'score')
south_asia_outliers_gdp = detect_outliers(south_asia_data, 'Log GDP per capita')

middle_east_outliers_score = detect_outliers(middle_east_data, 'score')
middle_east_outliers_gdp = detect_outliers(middle_east_data, 'Log GDP per capita')

# Combine outliers into a single dataframe for each region
south_asia_outliers = pd.concat([south_asia_outliers_score, south_asia_outliers_gdp]).drop_duplicates()
middle_east_outliers = pd.concat([middle_east_outliers_score, middle_east_outliers_gdp]).drop_duplicates()

# Print detected outliers
print("South Asia Outliers:")
print(south_asia_outliers[['Country name', 'score', 'Log GDP per capita']])

print("\nMiddle East Outliers:")
print(middle_east_outliers[['Country name', 'score', 'Log GDP per capita']])

# Visualization of outliers in a scatter plot
def plot_outliers(region_data, outliers, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(region_data['Log GDP per capita'], region_data['score'], label='Data Points', color='lightblue')
    plt.scatter(outliers['Log GDP per capita'], outliers['score'], label='Outliers', color='red')
    plt.title(title)
    plt.xlabel('GDP per Capita (Log)')
    plt.ylabel('score')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot for South Asia
plot_outliers(south_asia_data, south_asia_outliers, "South Asia: Outliers in Score and GDP per Capita")

# Plot for Middle East
plot_outliers(middle_east_data, middle_east_outliers, "Middle East: Outliers in Score and GDP per Capita")
```
