```python
def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):  # Check if the element is a list
            total += sum_nested_list(element)  # Recursively sum the nested list
        else:
            total += element  # Add the number to the total
    return total

nested_list_1 = [1, [2, [3, 4], 5], 6, [7, 8]]


print(sum_nested_list(nested_list_1))
```

```python
def generate_permutations(s):
    if len(s) == 1:
        return [s]

    # Recursive case
    permutations = []
    for i, char in enumerate(s):
        remaining = s[:i] + s[i+1:]
        for p in generate_permutations(remaining):
            # Concatenate the current character with the permutations of the remaining characters
            permutations.append(char + p)

    # Use set to remove duplicates and convert back to a list
    return list(set(permutations))

print(generate_permutations("anu"))
print(generate_permutations("ska"))
```

```python
def calculate_directory_size(directory):
    total_size = 0

    for item, value in directory.items():
        if isinstance(value, dict):
            total_size += calculate_directory_size(value)
        else:
            # If it's not a dictionary, it's a file, so add its size to the total
            total_size += value

    return total_size  # Return the total size

directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}


print(calculate_directory_size(directory_structure))
```

```python
def min_coins(coins, amount):
    """
    Finds the minimum number of coins needed to make up a given amount using dynamic programming.
    """
    # Initialize the DP table with a large value (infinity)
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0  # Base case: 0 coins are needed for amount 0

    # Fill the DP table
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

# Test case
coins = [1, 2, 5]
amount = 11
print(min_coins(coins, amount))  # Output: 3
```

```python
def longest_common_subsequence(s1, s2):
    """
    Finds the length of the Longest Common Subsequence (LCS) using dynamic programming.
    """
    n, m = len(s1), len(s2)
    dp = [[0] * (m + 1) for _ in range(n + 1)]

    # Fill the DP table
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[n][m]

# Test case
s1 = "abcde"
s2 = "ace"
print(longest_common_subsequence(s1, s2))
```

```python
def knapsack(weights, values, capacity):
    """
    Solves the 0/1 Knapsack problem using dynamic programming.
    """
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    # Fill the DP table
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]

# Testing
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
print(knapsack(weights, values, capacity))
```
